```python
final_df['Date_of_Journey'] = pd.to_datetime(final_df['Date_of_Journey'], format='%d/%m/%Y')
final_df['Year'] = final_df['Date_of_Journey'].dt.year
final_df['month'] = final_df['Date_of_Journey'].dt.month
final_df['date'] = final_df['Date_of_Journey'].dt.day
final_df.info()
```

```python
final_df.drop(['Date_of_Journey'],axis=1,inplace=True)
final_df.head()
```

```python
final_df['Arrival_Time'].value_counts()
```

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
final_df.head()
```

```python
final_df['Arrival_Time'] = pd.to_datetime(final_df['Arrival_Time'], format='%H:%M')
final_df['Arrival_hour'] = final_df['Arrival_Time'].dt.hour
final_df['Arrival_min'] = final_df['Arrival_Time'].dt.minute
final_df.drop(['Arrival_Time'],axis=1,inplace=True)
final_df.head()
```

```python
final_df['Dep_Time'] = pd.to_datetime(final_df['Dep_Time'], format='%H:%M')
final_df['Dep_Hour'] = final_df['Dep_Time'].dt.hour
final_df['Dep_min'] = final_df['Dep_Time'].dt.minute
final_df.drop(['Dep_Time'],axis=1,inplace=True)
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
final_df['Total_Stops'].value_counts()
```
